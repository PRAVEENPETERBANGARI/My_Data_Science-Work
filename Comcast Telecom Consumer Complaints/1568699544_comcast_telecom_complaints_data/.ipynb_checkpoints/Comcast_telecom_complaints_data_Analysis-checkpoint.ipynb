## Importing Necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 8]
import seaborn as sns
sns.set_style("whitegrid")
```

```python
data=pd.read_csv("Comcast_telecom_complaints_data.csv")
```

```python
data["Date_month_year"]=pd.to_datetime(data["Date_month_year"])
```

```python
data.dtypes
```

## Extracting month and day from Date column to get trend chart

```python
data["month"]=data["Date_month_year"].dt.month
data["day"]=data["Date_month_year"].dt.day
```

```python
data.head()
```

```python
month_group=data.groupby("month", as_index=False)["Customer Complaint"].count()
```

```python
month_group
```

## TREND CHART BASED ON MONTHLY COMPLAINTS

```python
sns.lineplot(data=month_group, x="month", y="Customer Complaint")
```

```python
daily_group=data.groupby("day", as_index=False)["Customer Complaint"].count()
```

```python
daily_group
```

## TREND CHART BASED ON DAILY COMPLAINTS

```python
sns.lineplot(data=daily_group, x="day", y="Customer Complaint")
```

#### Create a new categorical variable  with value as Open and Closed. Open & Pending is to be categorized as Open and Closed & Solved is to be categorized as Closed.

```python
conditions=[(data["Status"]=="Open") | (data["Status"]=="Pending"),
            (data["Status"]=="Closed") | (data["Status"]=="Solved")]
values=["Open", "Closed"]
data["New_Status"]=np.select(conditions, values)
```

```python
#data["Open"]="Not Open"
#data.loc[(data["Status"]=="Open") | (data["Status"]=="Pending"), "Open"]= "Open"
```

```python
data["New_Status"].value_counts().plot(kind="bar")
```

## state wise status of complaints in a stacked bar chart

```python
#state_group=data.groupby(["State", "New_Status"], as_index=False).count()
```

```python
state_group=data.groupby(["State", "New_Status"]).size().reset_index().pivot(columns='New_Status', index='State', values=0)
```

```python
state_group.plot(kind='bar', stacked=True)
```

```python
state_group.sort_values(by="Closed", ascending=False)
```

## From the stacked bar and above we can say that *Georgia* state has the maximum complaints
